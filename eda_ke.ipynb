```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 随机分组
np.random.seed(42)  # 保证可复现
df['group'] = np.random.randint(0, 10, size=len(df))

# 收集 prevtrandate 列
prevdate_cols = [f'prevtrandate{i}' for i in range(1, 11)]

# 转成长格式
df_long = df[['group'] + prevdate_cols].melt(id_vars='group', 
                                              var_name='prevdate_index', 
                                              value_name='days_before')

# 保证是整数/float
df_long['days_before'] = pd.to_numeric(df_long['days_before'], errors='coerce')
df_long['prevdate_index'] = df_long['prevdate_index'].str.extract('(\d+)').astype(int)

# 按组和交易位置计算均值
grouped_mean = df_long.groupby(['group', 'prevdate_index'])['days_before'].mean().reset_index()

# 画图
plt.figure(figsize=(12, 6))
for g in grouped_mean['group'].unique():
    subset = grouped_mean[grouped_mean['group'] == g]
    plt.plot(subset['prevdate_index'], subset['days_before'], marker='o', label=f'Group {g}')

plt.xlabel("Prevtrandate Index (1 = Most Recent)")
plt.ylabel("Average Days Before Deposit")
plt.title("Average Days Before Deposit by Prevtrandate Index (Grouped)")
plt.gca().invert_xaxis()  # 让最近的交易在右边
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/kzhan176/citizens_data/BrownDSI_masked_capstone_data.csv_20250401031515')

prevtran_cols = [f'prevtran{i}' for i in range(1, 11)]
prevtrandate_cols = [f'prevtrandate{i}' for i in range(1, 11)]

prevtran_melted = df[prevtran_cols].melt(value_name='prevtran')
prevtrandate_melted = df[prevtrandate_cols].melt(value_name='prevtrandate')

# Plot distributions
plt.figure(figsize=(10, 5))
sns.histplot(prevtran_melted['prevtran'].dropna(), kde=True)
plt.title('Distribution of prevtran values')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(prevtrandate_melted['prevtrandate'].dropna(), kde=True)
plt.title('Distribution of prevtrandate (days ago)')
plt.show()
```
